```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open('newyear2018.txt', mode='rt', encoding = 'utf-8').read()
text
```

```python
import nltk
```

```python
from konlpy.tag import Okt; okt=Okt()
tokens = okt.nouns(text)
tokens
```

```python
newyear = nltk.Text(tokens, name='신년사')
```

```python
print(len(newyear.tokens))
print(len(set(newyear.tokens)))
newyear.vocab()
```

```python
import matplotlib.pyplot as plt
import platform
%matplotlib inline

from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
    
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(12,6))
newyear.plot(50)
plt.show()
```

```python
stop_words = ['.',',','의','것','수','위','이','가','등','번','을','를','만','더','로','저','위',
              '다','전','그','일','차','길','틀','곳','또','위해','해','올해','년','지난','여러분']
newyear = [each_word for each_word in newyear if each_word not in stop_words]
newyear
```

```python
newyear = nltk.Text(newyear, name='신년사')
plt.figure(figsize=(12, 6))
newyear.plot(50)
plt.show()
```

```python
newyear.count('평화')
```

```python
plt.figure(figsize=(12,6))
newyear.dispersion_plot(['정부','평화','경제'])
```

```python
newyear.concordance('평화')
```

```python
korea_mask = np.array(Image.open('korea_layout.png'))
```

```python
plt.figure(figsize=(8,8))
plt.imshow(korea_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
data = newyear.vocab().most_common(100)
wordcloud = WordCloud(font_path='/Library/Fonts/AppleGothic.ttf',
                      relative_scaling = 0.2, background_color='white', mask=korea_mask,
                     ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show
```
